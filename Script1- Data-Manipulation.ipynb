#### UNPIVOT

```python
id_vars = list(data_modified.columns)[:8]
value_vars = list(data_modified.columns)[8:]
dataset_melted = data_modified.melt(id_vars=id_vars, value_vars=value_vars,var_name="Question + Subquestion",value_name="Answer")
```

```python
dataset_melted
```

```python
question_import = pd.read_excel(pwd + "//Data - Survey Monkey Edited.xlsx",sheet_name= "Question")
```

```python
question_import
```

```python
question_import.drop(columns=["Raw Question","Raw sub-questions"],inplace=True)
```

```python
question_import.dropna(inplace=True)
```

```python
question=question_import.copy
```

```python
question
```

#### Join the two imported file

```python
dataset_merged =pd.merge(left=dataset_melted,right=question_import,how="left",left_on="Question + Subquestion",right_on="Question + Subquestion")
```

```python
print("Original Data", len(dataset_melted))
print("Merged Data", len(dataset_merged))
dataset_merged
```
